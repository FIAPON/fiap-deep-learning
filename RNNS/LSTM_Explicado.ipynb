```python
import numpy as np

# Função de ativação sigmóide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Função de ativação tangente hiperbólica
def tanh(x):
    return np.tanh(x)

# Entrada: frase simples dividida em palavras
frases = ["A", "LSTM", "é", "uma", "rede", "recorrente"]


# Inicialização das variáveis da LSTM
hidden_state = 0  # Estado oculto inicial
cell_state = 0    # Estado da célula inicial

# Parâmetros da LSTM (valores fictícios para fins didáticos)
input_gate_weights = np.random.rand(6)
forget_gate_weights = np.random.rand(6)
output_gate_weights = np.random.rand(6)
cell_weights = np.random.rand(6)

# Lista para armazenar as palavras que contribuem para o cell_state em cada etapa
palavras_contribuintes = []

# Simulação do processamento da frase
for palavra in frases:
    # Representação da palavra como vetor de entrada (simplificado)
    word_vector = np.random.rand(6)

    # Cálculo do input_gate
    input_gate = sigmoid(np.dot(input_gate_weights, word_vector))

    # Cálculo do forget_gate
    forget_gate = sigmoid(np.dot(forget_gate_weights, word_vector))

    # Cálculo do output_gate
    output_gate = sigmoid(np.dot(output_gate_weights, word_vector))

    # Cálculo do cell_state_candidate
    cell_state_candidate = tanh(np.dot(cell_weights, word_vector))

    # Atualização do cell_state
    cell_state = forget_gate * cell_state + input_gate * cell_state_candidate

    # Atualização do hidden_state
    hidden_state = output_gate * tanh(cell_state)

    # Armazena a palavra contribuinte para o cell_state_candidate
    palavras_contribuintes.append((palavra, cell_state_candidate))

    # Exibição dos resultados
    print(f"Palavra: {palavra}")
    print(f"Input Gate: {input_gate}")
    print(f"Forget Gate: {forget_gate}")
    print(f"Output Gate: {output_gate}")
    print(f"Cell State Candidate: {cell_state_candidate}")
    print(f"Cell State: {cell_state}")
    print(f"Hidden State: {hidden_state}")
    print("\n")

# Resultado final após processar todas as palavras
print("Resultado Final:")
print(f"Cell State Final: {cell_state}")
print(f"Hidden State Final: {hidden_state}")

```
